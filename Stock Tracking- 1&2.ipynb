# Backtesting a momentum strategy

### Some basic investment terms

Positions to take on a stock:

+1: investing (long Position), you gain if prices rise and loose if prices fall

-1: short selling (short position) opposite of investing, you gain if prices fall

0: No position (neutral)

### Strategies

1) Buy and Hold (Basic strategy: passive), Position: +1 on any given day

2) Simple Momentum (active strategy); There are two strategies: 

    a) If today's return was positive, then we would position +1 tomorrow

    b) If today's return was negative, then we would position -1 tomorrow

```python
# Positions to take on a stock:

df['Position'] = np.sign(df.DJI_Return)
df
```

```python
# Return

df['Strategy_Return'] = df.Position.shift() * df.DJI_Return
df
```

```python
# calculating the cumulative investment multiple

df.Strategy_Return.add(1, fill_value = 0).cumprod()
```

```python
# we multiply all these 'cumulative investment multiple' by the first investment (DJI_Close = 12480.69)

df['Strategy'] = df.Strategy_Return.add(1, fill_value = 0).cumprod() * df.iloc[0,0]
df
```

```python
df.iloc[0,0]
```

```python
# Return that we would have obtained
df_plot = df.drop(['DJI_Close', 'DJI_Return', 'Position', 'Strategy_Return'], axis = 1)
df_plot
```

```python
sns.set()
df_plot.plot(figsize = (20, 12), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
# Comparing buy and hold strategy with momentum strategy

sns.set()
df[['DJI_Close', 'Strategy']].plot(figsize = (16, 10), fontsize = 16)
plt.title('Buy and hold strategy vs. Momentum strategy', fontsize = 18)
plt.legend(fontsize = 16)
plt.show()
```

```python
# Calculating annualized 'risk' and 'return' for both the strategies

def summary_ann(returns):
    summary = returns.agg(['mean', 'std']).T
    summary['Return'] = summary['mean'] * 252  # because we have 252 trading days in a year
    summary['Risk'] = summary['std'] * np.sqrt(252)
    summary.drop(columns = ['mean', 'std'],inplace = True)
    return summary
    
```

```python
summary_ann(df[['DJI_Return', 'Strategy_Return']])

# Clearly, the 'simple momentum' strategy is inferior to the basic 'buy and hold' strategy
```

## Backtesting vs fitting

Making investment strategies based on historical trends and data (fitting astrategy
based on historical prices) is one if the poorest strategies, since they do not say anything about the future profitability.

Instead, we nedd to test the strategies on new/future data, this is called "Forward Testing".

This is similar to ML, hence you cannot test the strategy on training data, you need fresh data(Test set).

```python
dji
```

```python
# We are using our previous contrarian strategy
# However we implement a general code, that also allows for future flexibility

df = dji.loc["2010":,'Close'].to_frame()
df['DJI_Return'] = df.Close.pct_change()
df.columns = ['DJI_Close', 'DJI_Return']
df.dropna(inplace = True)
df
```

```python
# Contrarian strategy in a generalized flexible form
df['Position'] = np.where(df.DJI_Return > 0, -1, 1) 
df
```

```python
df['Strategy_Return'] = df['Position'].shift() * df['DJI_Return']
df['Strategy'] = df.Strategy_Return.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
df
```

```python
sns.set()
df[['DJI_Close', 'Strategy']].plot(figsize = (16, 10), fontsize = 15)
plt.title('Something', fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
summary_ann(df[['DJI_Return', 'Strategy_Return']])
```

### Let us tweak our backtesting strategy a bit

Instead of shortselling the stock when it goes positive we rather shortsell it only when it returns more than 1% in a day

```python

dji
```

```python
# We are using our previous contrarian strategy
# However we implement a general code, that also allows for future flexibility

df_tweaked = dji.loc["2010":,'Close'].to_frame()
df_tweaked['DJI_Return'] = df_tweaked.Close.pct_change()
df_tweaked.columns = ['DJI_Close', 'DJI_Return']
df_tweaked.dropna(inplace = True)
df_tweaked
```

```python
# Contrarian strategy for when stocks  go greater than 1% (in a day)
df_tweaked['Position'] = np.where(df_tweaked.DJI_Return > 0.01, -1, 1) 
df_tweaked
```

```python
df_tweaked['Strategy_Return'] = df_tweaked['Position'].shift() * df_tweaked['DJI_Return']
df_tweaked['Strategy'] = df_tweaked.Strategy_Return.add(1, fill_value = 0).cumprod() * df_tweaked.iloc[0, 0]
df_tweaked
```

```python
sns.set()
df_tweaked[['DJI_Close', 'Strategy']].plot(figsize = (16, 10), fontsize = 15)
plt.title('Something', fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
summary_ann(df_tweaked[['DJI_Return', 'Strategy_Return']])

# With our new tweak we get a return of 17%

# We can try different values for backtesting and choose the best one

# Also we do not rely only on backtesting, we also do forward testing on this
```

# Strategies based on Simple Moving Averages (SMA) 

These strategies are also called 'crossovers', we would see why?

### SMA

```python
dji
```

```python
df = dji.loc['2010':, 'Close'].to_frame()
df['DJI_Return'] = df.Close.pct_change()
df.columns = ['DJI_Close', 'DJI_Return']
df.dropna(inplace = True)
df
```

```python
# smoothening the change in prices by calculating moving averages (over a period of 50 days)
df.DJI_Close.rolling(window = 50).mean()
```

```python
df['SMA_50'] = df.DJI_Close.rolling(window = 50).mean()
```

```python
df[['DJI_Close', 'SMA_50']].plot(figsize = (14, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()

# As we said, it smoothens the curve
```

```python
# Similarly we calculate moving averages (over a period of 200 days)
df['SMA_200'] = df.DJI_Close.rolling(window = 200).mean()
```

```python
df[['SMA_50', 'SMA_200']].plot(figsize = (14, 10), fontsize = 15)
plt.title('Comparing the two SMAs', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()

# SMA_200 smoothens the curve even further

# SMA_50 however, reacts faster to the change on trends
```

#### As we see, there are points where the curves cross eachother; These are what we call "Moving Average Crossovers" and we will base our strategies on these

## Now we implement the SMA Crossover momentum strategy 

#### a) Investing (+1): SMA50 > SMA200
#### b) short Selling (-1): SMA50 < SMA200

```python
df.dropna(inplace = True)
df['Position'] = np.sign(df['SMA_50'].sub(df['SMA_200']))
df
```

```python
df[['SMA_50', 'SMA_200', 'Position']].plot(figsize = (14, 10), fontsize = 15, secondary_y = 'Position')
plt.show()
```

```python
df['Strategy_Return'] = df['Position'].shift() * df['DJI_Return']
df
```

```python
df['Strategy'] = df['Strategy_Return'].add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
df
```

```python
df[['DJI_Close', 'Strategy']].plot(figsize = (14, 10), fontsize = 15)
plt.title('SMA Crossover momentum strategy')
plt.legend(fontsize = 15)
plt.show()
```

```python
summary_ann(df[['DJI_Return', 'Strategy_Return']])
```
